```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_data = CORE_DATA_DIR + '/input_data/azure/5_mins.csv'
df = pd.read_csv(file_data, header=None)
```

```python
def plot_data(df, column_index, figsize=(25, 20), title=None):
    df1 = df.iloc[:, column_index]
    df1.plot(linewidth=0.5, figsize=figsize, title=title)
    plt.show()
```

```python
plot_data(df, 1, title='Azure: 0 -> 8833')
```

```python
plot_data(df[:1000], 1, title='Azure: 0 -> 1000')
```

```python
plot_data(df[:100], 1, title='Azure: 0 -> 100')
```
